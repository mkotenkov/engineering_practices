```python
import pandas as pd

DATA_PATH = "../data/raw/data.csv"
```

# Провести такое же описание пассажиров отдельно для каждого класса билета. Составить средний портрет пассажира каждого класса

```python
df = pd.read_csv(DATA_PATH)
df.drop(["PassengerId"], axis=1, inplace=True)
df.head()
```

```python
age_groups = [
    (0, 14),
    (14, 21),
    (21, 35),
    (35, 55),
    (55, 200),
]

for age_group in age_groups:
    print("AGE GROUP:", age_group)
    
    filtered_df = df[(df["Age"] >= age_group[0]) & (df["Age"] < age_group[1])]

    print("Mean age:", round(filtered_df["Age"].mean(), 2))
    print("Mean fare:", round(filtered_df["Fare"].mean(), 2))
    print("Mean SibSp:", round(filtered_df["SibSp"].mean(), 2))
    print("Mean Parch:", round(filtered_df["Parch"].mean(), 2))
    print("Survival prob:", round(filtered_df["Survived"].mean(), 2))
    print("Male/ female ratio:", round(filtered_df[filtered_df["Sex"] == "male"].shape[0] / filtered_df.shape[0], 2))
    print("Most common Embarked:", filtered_df["Embarked"].value_counts().idxmax())
    print("Most common Cabin:", filtered_df["Cabin"].value_counts().idxmax())

    print()
```
